# Write a code to count the number of vowels in a string

```python
name = "Aditi Patil"
vowels = "aeiouAEIOU"
vowel_count = 0
for char in name:
  if char in vowels:
    vowel_count = vowel_count + 1
print(F"Number of Vowels in {name}: {vowel_count}")
```

# Write a code to find all occurrences of a given substring within another string

```python
s = "hello world, hello universe"
substring = "hello"
positions = [i for i in range(len(s)) if s.startswith(substring, i)]

print(positions)
```
